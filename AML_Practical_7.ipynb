7. Least Squares Estimation:

**Scenario Based Question:**
A researcher wants to study the relationship between advertising expenditure (in thousands of dollars) and sales revenue (in thousands of dollars). They collect data from 8 companies. Using the Least Squares Estimation (LSE) method:
--Estimate the regression coefficients (β₀, β₁).
--Fit a regression line to the data.
--Predict sales revenue for a new company that spends $15k on advertising.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
advertising = np.array([5, 7, 8, 10, 12, 15, 18, 20])   # Advertising spend ($k)
sales = np.array([14, 17, 20, 23, 26, 30, 35, 40])     # Sales revenue ($k)
```

```python
# Least Squares Estimation formula
n = len(advertising)
x_mean = np.mean(advertising)
y_mean = np.mean(sales)
```

```python
# Calculate slope (β1) and intercept (β0)
beta1 = np.sum((advertising - x_mean) * (sales - y_mean)) / np.sum((advertising - x_mean)**2)
beta0 = y_mean - beta1 * x_mean
```

```python
print("Intercept (β0):", round(beta0, 3))
print("Slope (β1):", round(beta1, 3))
```

```python
# Regression line
y_pred = beta0 + beta1 * advertising
```

```python
# Prediction for new advertising spend = $15k
new_x = 15
new_y = beta0 + beta1 * new_x
print(f"Predicted Sales for $15k advertising: {round(new_y, 2)}k")
```

```python
# Plot
plt.scatter(advertising, sales, color="blue", label="Observed Data")
plt.plot(advertising, y_pred, color="red", label="Regression Line")
plt.scatter(new_x, new_y, color="green", marker="o", s=100, label="New Prediction")
plt.xlabel("Advertising Spend ($k)")
plt.ylabel("Sales Revenue ($k)")
plt.title("Least Squares Estimation - Regression Line")
plt.legend()
plt.show()
```
