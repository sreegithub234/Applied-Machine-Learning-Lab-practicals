2. Evaluation of Simple Linear Regression

**Scenario-based Question:**
A real estate analyst wants to predict house prices based on area in square feet. Fit a simple linear regression model and evaluate its performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Dataset
area = np.array([800, 1000, 1200, 1500, 1800, 2000]).reshape(-1,1)
price = np.array([150, 180, 200, 230, 260, 290])  # in lakhs
```

```python
# Train Model
model = LinearRegression()
model.fit(area, price)
```

```python
# Predictions
predicted_price = model.predict(area)
```

```python
# Plot
plt.scatter(area, price, color="blue", label="Actual Data")
plt.plot(area, predicted_price, color="red", label="Regression Line")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (Lakhs)")
plt.legend()
plt.show()
```
